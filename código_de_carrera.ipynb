```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_eficiencia(produccion_real, produccion_ideal):
    """
    Calcula la eficiencia de una estaci贸n de trabajo en una l铆nea de producci贸n.
    """
    if produccion_ideal == 0:
        return 0
    return (produccion_real / produccion_ideal) * 100


estaciones = ["Corte", "Ensamble", "Pintura", "Empaque"]
produccion_real = []
produccion_ideal = []


print(" Ingrese los datos de producci贸n para cada estaci贸n ")
for estacion in estaciones:
    real = int(input(f"Producci贸n real en {estacion}: "))
    ideal = int(input(f"Producci贸n ideal en {estacion}: "))

    produccion_real.append(real)
    produccion_ideal.append(ideal)


eficiencia = [calcular_eficiencia(real, ideal) for real, ideal in zip(produccion_real, produccion_ideal)]


print("\n Resultados de eficiencia ")
for i in range(len(estaciones)):
    print(f"{estaciones[i]} - Eficiencia: {eficiencia[i]:.2f}%")


plt.figure(figsize=(8,5))
x = np.arange(len(estaciones))
plt.bar(x - 0.2, produccion_real, width=0.4, label="Producci贸n Real", color="blue")
plt.bar(x + 0.2, produccion_ideal, width=0.4, label="Producci贸n Ideal", color="orange")

plt.xticks(x, estaciones)
plt.ylabel("Unidades Producidas")
plt.title("Comparaci贸n de Producci贸n Real vs Ideal")
plt.legend()
plt.show()
```
