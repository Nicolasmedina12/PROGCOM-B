```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_eficiencia(produccion_real, produccion_ideal):
    """
    Calcula la eficiencia de una estación de trabajo en una línea de producción.
    """
    if produccion_ideal == 0:
        return 0
    return (produccion_real / produccion_ideal) * 100


estaciones = ["Corte", "Ensamble", "Pintura", "Empaque"]
produccion_real = []
produccion_ideal = []


print("🔹 Ingrese los datos de producción para cada estación 🔹")
for estacion in estaciones:
    real = int(input(f"Producción real en {estacion}: "))
    ideal = int(input(f"Producción ideal en {estacion}: "))

    produccion_real.append(real)
    produccion_ideal.append(ideal)


eficiencia = [calcular_eficiencia(real, ideal) for real, ideal in zip(produccion_real, produccion_ideal)]


print("\n📊 Resultados de eficiencia 📊")
for i in range(len(estaciones)):
    print(f"{estaciones[i]} - Eficiencia: {eficiencia[i]:.2f}%")


plt.figure(figsize=(8,5))
x = np.arange(len(estaciones))
plt.bar(x - 0.2, produccion_real, width=0.4, label="Producción Real", color="blue")
plt.bar(x + 0.2, produccion_ideal, width=0.4, label="Producción Ideal", color="orange")

plt.xticks(x, estaciones)
plt.ylabel("Unidades Producidas")
plt.title("Comparación de Producción Real vs Ideal")
plt.legend()
plt.show()
```
