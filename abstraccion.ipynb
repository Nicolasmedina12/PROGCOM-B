```python
import random

# Ejercicio 1: Clase Saludo
class Saludo:
    def saludar(self, nombre):
        print(f"Hola, {nombre}")

# Ejercicio 2: Clase Numero
class Numero:
    def __init__(self, valor):
        self.valor = valor

    def es_par(self):
        return self.valor % 2 == 0

# Ejercicio 3: Clase Rectangulo
class Rectangulo:
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura

# Ejercicio 4: Clase Temperatura
class Temperatura:
    def __init__(self, celsius):
        self.celsius = celsius

    def a_fahrenheit(self):
        return self.celsius * 1.8 + 32

# Ejercicio 5: Clase Palabra
class Palabra:
    def __init__(self, texto):
        self.texto = texto.lower()

    def contar_vocales(self):
        return sum(1 for c in self.texto if c in "aeiou")

# Ejercicio 6: Clase Palindromo
class Palindromo:
    def __init__(self, palabra):
        self.palabra = palabra.lower()

    def es_palindromo(self):
        return self.palabra == self.palabra[::-1]

# Ejercicio 7: Clase Dado
class Dado:
    def tirar(self):
        return random.randint(1, 6)

# Ejercicio 8: Clase ListaCuadrados
class ListaCuadrados:
    def generar(self, n):
        return [i**2 for i in range(1, n+1)]

# Ejercicio 9: Clase Calculadora
class Calculadora:
    def operar(self, a, b, tipo):
        if tipo == "suma":
            return a + b
        elif tipo == "resta":
            return a - b
        elif tipo == "multiplicacion":
            return a * b
        elif tipo == "division":
            return a / b if b != 0 else "Error: división por cero"
        else:
            return "Operación no válida"

# Ejercicio 10: Clase GeneradorContrasena
class GeneradorContrasena:
    def crear(self, nombre, año):
        return f"{nombre}@{año}"
```

```python
import java.util.Random;

class Saludo {
    public void saludar(String nombre) {
        System.out.println("Hola, " + nombre);
    }
}

class Numero {
    private int valor;

    public Numero(int valor) {
        this.valor = valor;
    }

    public boolean esPar() {
        return valor % 2 == 0;
    }
}

class Rectangulo {
    private int base;
    private int altura;

    public Rectangulo(int base, int altura) {
        this.base = base;
        this.altura = altura;
    }

    public int area() {
        return base * altura;
    }
}

class Temperatura {
    private double celsius;

    public Temperatura(double celsius) {
        this.celsius = celsius;
    }

    public double aFahrenheit() {
        return celsius * 1.8 + 32;
    }
}

class Palabra {
    private String texto;

    public Palabra(String texto) {
        this.texto = texto.toLowerCase();
    }

    public int contarVocales() {
        int count = 0;
        for (char c : texto.toCharArray()) {
            if ("aeiou".indexOf(c) >= 0) {
                count++;
            }
        }
        return count;
    }
}

class Palindromo {
    private String palabra;

    public Palindromo(String palabra) {
        this.palabra = palabra.toLowerCase();
    }

    public boolean esPalindromo() {
        String reversa = new StringBuilder(palabra).reverse().toString();
        return palabra.equals(reversa);
    }
}

class Dado {
    private Random random = new Random();

    public int tirar() {
        return random.nextInt(6) + 1;
    }
}

class ListaCuadrados {
    public int[] generar(int n) {
        int[] cuadrados = new int[n];
        for (int i = 0; i < n; i++) {
            cuadrados[i] = (i + 1) * (i + 1);
        }
        return cuadrados;
    }
}

class Calculadora {
    public double operar(double a, double b, String tipo) {
        switch (tipo) {
            case "suma": return a + b;
            case "resta": return a - b;
            case "multiplicacion": return a * b;
            case "division": return b != 0 ? a / b : Double.NaN;
            default: throw new IllegalArgumentException("Operación no válida");
        }
    }
}

class GeneradorContrasena {
    public String crear(String nombre, int anio) {
        return nombre + "@" + anio;
    }
}
```
