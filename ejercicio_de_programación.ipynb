```python
import random
import pandas as pd

# Generar nombres aleatorios
nombres = [
    "Juan", "María", "Carlos", "Ana", "Luis", "Camila", "Jorge", "Sofía", "Andrés", "Laura",
    "Pedro", "Daniela", "Santiago", "Valentina", "Diego", "Paula", "Felipe", "Manuela", "David", "Isabella"
]

def generar_persona():
    nombre = f"{random.choice(nombres)} {random.choice(['García', 'Rodríguez', 'Martínez', 'López'])}"
    telefono = f"+57 {random.choice(['300', '301', '310', '311'])} {random.randint(1000000,9999999)}"
    peso = round(random.uniform(45, 120), 1)
    altura = round(random.uniform(1.5, 2.0), 2)
    imc = round(peso / (altura ** 2), 2)
    ritmo = random.randint(55, 120)

    if imc < 18.5:
        plan = "Aumentar masa muscular"
    elif imc < 25:
        plan = "Mantener estado físico"
    elif imc < 30:
        plan = "Cardio y fuerza moderada"
    else:
        plan = "Cardio intenso y pérdida de peso"

    return {
        "nombre": nombre,
        "telefono": telefono,
        "peso_kg": peso,
        "altura_m": altura,
        "imc": imc,
        "ritmo_cardiaco_bpm": ritmo,
        "plan_entrenamiento": plan
    }

# Generar lista de personas
clientes = [generar_persona() for _ in range(10)]
df = pd.DataFrame(clientes)

# Función para modificar datos según el rol
def modificar_cliente(df, nombre_cliente, rol="cliente", **kwargs):
    idx = df[df["nombre"] == nombre_cliente].index
    if idx.empty:
        print("Cliente no encontrado.")
        return df
    i = idx[0]

    # Permisos según rol
    permisos = {
        "cliente": {"nombre", "telefono"},
        "entrenador": {"peso_kg", "ritmo_cardiaco_bpm", "plan_entrenamiento"},
        "admin": set(df.columns)
    }

    campos_permitidos = permisos.get(rol)
    if campos_permitidos is None:
        print("Rol no válido.")
        return df

    for campo, valor in kwargs.items():
        if campo in campos_permitidos:
            df.at[i, campo] = valor
            if campo == "peso_kg":
                altura = df.at[i, 'altura_m']
                df.at[i, 'imc'] = round(valor / (altura ** 2), 2)  # Recalcular IMC
        else:
            print(f"No puedes modificar el campo '{campo}' con el rol '{rol}'")

    return df

# Función para mostrar datos visibles según rol
def ver_datos(df, rol="cliente"):
    permisos_visibles = {
        "cliente": ["nombre", "telefono"],
        "entrenador": ["nombre", "peso_kg", "altura_m", "imc", "ritmo_cardiaco_bpm", "plan_entrenamiento"],
        "admin": list(df.columns)
    }
    columnas = permisos_visibles.get(rol, [])
    return df[columnas]

# === DEMOSTRACIÓN ===
print("=== VISTA DEL CLIENTE ===")
print(ver_datos(df, rol="cliente"))

print("\n=== CLIENTE CAMBIA SU TELÉFONO ===")
df = modificar_cliente(df, nombre_cliente="Juan García", rol="cliente", telefono="+57 300 1234567")

print("\n=== VISTA DEL ENTRENADOR ===")
print(ver_datos(df, rol="entrenador"))

print("\n=== ADMIN CAMBIA TODO ===")
df = modificar_cliente(df, nombre_cliente="Juan García", rol="admin", altura_m=1.90, nombre="Juan M. García", peso_kg=85)

print("\n=== DATOS COMPLETOS PARA ADMIN ===")
print(ver_datos(df, rol="admin"))
```

```python
import random
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# Generar datos
nombres = [
    "Juan", "María", "Carlos", "Ana", "Luis", "Camila", "Jorge", "Sofía", "Andrés", "Laura",
    "Pedro", "Daniela", "Santiago", "Valentina", "Diego", "Paula", "Felipe", "Manuela", "David", "Isabella"
]

def generar_persona():
    nombre = f"{random.choice(nombres)} {random.choice(['García', 'Rodríguez', 'Martínez', 'López'])}"
    telefono = f"+57 {random.choice(['300', '301', '310', '311'])} {random.randint(1000000,9999999)}"
    peso = round(random.uniform(45, 120), 1)
    altura = round(random.uniform(1.5, 2.0), 2)
    imc = round(peso / (altura ** 2), 2)
    ritmo = random.randint(55, 120)

    if imc < 18.5:
        plan = "Aumentar masa muscular"
    elif imc < 25:
        plan = "Mantener estado físico"
    elif imc < 30:
        plan = "Cardio y fuerza moderada"
    else:
        plan = "Cardio intenso y pérdida de peso"

    return {
        "nombre": nombre,
        "telefono": telefono,
        "peso_kg": peso,
        "altura_m": altura,
        "imc": imc,
        "ritmo_cardiaco_bpm": ritmo,
        "plan_entrenamiento": plan
    }

# Crear DataFrame
clientes = [generar_persona() for _ in range(10)]
df = pd.DataFrame(clientes)

# Permisos por rol
permisos_visibles = {
    "cliente": ["nombre", "telefono"],
    "entrenador": ["nombre", "peso_kg", "altura_m", "imc", "ritmo_cardiaco_bpm", "plan_entrenamiento"],
    "admin": list(df.columns)
}

permisos_edicion = {
    "cliente": {"nombre", "telefono"},
    "entrenador": {"peso_kg", "ritmo_cardiaco_bpm", "plan_entrenamiento"},
    "admin": set(df.columns)
}

# === Widgets ===
rol_widget = widgets.Dropdown(options=["cliente", "entrenador", "admin"], description="Rol:")
nombre_widget = widgets.Dropdown(options=df["nombre"].tolist(), description="Cliente:")

# Campos editables
telefono_widget = widgets.Text(description="Teléfono:")
peso_widget = widgets.FloatText(description="Peso (kg):")
ritmo_widget = widgets.IntText(description="Ritmo (bpm):")
plan_widget = widgets.Text(description="Plan:")
nombre_edit_widget = widgets.Text(description="Nombre:")
altura_widget = widgets.FloatText(description="Altura (m):")

btn_actualizar = widgets.Button(description="Actualizar")
output = widgets.Output()

def actualizar_vista(*args):
    output.clear_output()
    rol = rol_widget.value
    nombre = nombre_widget.value
    cliente = df[df["nombre"] == nombre].iloc[0]

    with output:
        print(f"== Vista para rol: {rol.upper()} ==")
        display(cliente[permisos_visibles[rol]])

    # Activar widgets según permisos
    nombre_edit_widget.value = cliente["nombre"]
    telefono_widget.value = cliente["telefono"]
    peso_widget.value = cliente["peso_kg"]
    ritmo_widget.value = cliente["ritmo_cardiaco_bpm"]
    plan_widget.value = cliente["plan_entrenamiento"]
    altura_widget.value = cliente["altura_m"]

    nombre_edit_widget.disabled = "nombre" not in permisos_edicion[rol]
    telefono_widget.disabled = "telefono" not in permisos_edicion[rol]
    peso_widget.disabled = "peso_kg" not in permisos_edicion[rol]
    ritmo_widget.disabled = "ritmo_cardiaco_bpm" not in permisos_edicion[rol]
    plan_widget.disabled = "plan_entrenamiento" not in permisos_edicion[rol]
    altura_widget.disabled = "altura_m" not in permisos_edicion[rol]

def aplicar_cambios(b):
    rol = rol_widget.value
    nombre = nombre_widget.value
    idx = df[df["nombre"] == nombre].index[0]

    # Solo aplicar si el campo está permitido
    if "nombre" in permisos_edicion[rol]:
        df.at[idx, "nombre"] = nombre_edit_widget.value
    if "telefono" in permisos_edicion[rol]:
        df.at[idx, "telefono"] = telefono_widget.value
    if "peso_kg" in permisos_edicion[rol]:
        df.at[idx, "peso_kg"] = peso_widget.value
        altura = df.at[idx, "altura_m"]
        df.at[idx, "imc"] = round(peso_widget.value / (altura ** 2), 2)
    if "altura_m" in permisos_edicion[rol]:
        df.at[idx, "altura_m"] = altura_widget.value
        peso = df.at[idx, "peso_kg"]
        df.at[idx, "imc"] = round(peso / (altura_widget.value ** 2), 2)
    if "ritmo_cardiaco_bpm" in permisos_edicion[rol]:
        df.at[idx, "ritmo_cardiaco_bpm"] = ritmo_widget.value
    if "plan_entrenamiento" in permisos_edicion[rol]:
        df.at[idx, "plan_entrenamiento"] = plan_widget.value

    nombre_widget.options = df["nombre"].tolist()
    actualizar_vista()

# Vínculos
rol_widget.observe(actualizar_vista, names='value')
nombre_widget.observe(actualizar_vista, names='value')
btn_actualizar.on_click(aplicar_cambios)

# Mostrar
display(rol_widget, nombre_widget, nombre_edit_widget, telefono_widget,
        peso_widget, altura_widget, ritmo_widget, plan_widget, btn_actualizar, output)

# Cargar vista inicial
actualizar_vista()
```
