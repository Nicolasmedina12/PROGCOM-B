False → Representa el valor booleano falso (0).

None → Representa la ausencia de un valor o un objeto nulo.

True → Representa el valor booleano verdadero (1).

and → Operador lógico AND (devuelve True si ambas condiciones son verdaderas).

or → Operador lógico OR (devuelve True si al menos una condición es verdadera).

not → Operador lógico NOT (invierte el valor de verdad).

is → Compara si dos variables apuntan al mismo objeto en memoria.

in → Verifica si un elemento está dentro de una lista, tupla, string, etc.

if → Inicia una estructura condicional.

elif → Define una condición alternativa en una estructura if.

else → Bloque de código que se ejecuta si if y elif no se cumplen.

while → Inicia un bucle que se repite mientras una condición sea True.

for → Inicia un bucle que itera sobre una secuencia (listas, tuplas, cadenas, etc.).

break → Detiene un bucle for o while antes de que termine.

continue → Salta a la siguiente iteración de un bucle sin ejecutar el código restante.

pass → Se usa como un marcador de posición para código que aún no se ha implementado.

def → Se usa para definir una función.

return → Devuelve un valor desde una función.

lambda → Define una función anónima en una sola línea.

class → Se usa para definir una clase (POO).

try → Inicia un bloque de código donde pueden ocurrir excepciones.

except → Captura una excepción y maneja el error.

finally → Bloque opcional que se ejecuta después de try-except, siempre, haya error o no.

raise → Lanza una excepción personalizada.

assert → Verifica si una condición es True, si no, lanza un error.

with → Gestiona el uso de recursos como archivos, cerrándolos automáticamente.

as → Se usa para renombrar módulos o manejar alias en with.

del → Elimina una variable, objeto o elemento de una lista o diccionario.

global → Declara una variable como global dentro de una función.

nonlocal → Modifica una variable en un ámbito no global, pero superior al actual.

async → Declara una función asincrónica que se ejecuta en segundo plano.

await → Pausa la ejecución de una función async hasta que termine otra tarea asincrónica.

yield → Permite que una función actúe como un generador, devolviendo valores sin perder su estado.

import → Se usa para importar módulos en un programa.

from → Se usa con import para importar partes específicas de un módulo.

int → Tipo de dato para números enteros.

float → Tipo de dato para números decimales.

str → Tipo de dato para cadenas de texto.

list → Tipo de dato para listas (estructuras mutables y ordenadas).

tuple → Tipo de dato para tuplas (estructuras inmutables y ordenadas).

dict → Tipo de dato para diccionarios (estructura clave-valor).

set → Tipo de dato para conjuntos (colección única de elementos).

bool → Tipo de dato booleano (True o False).

bytes → Representa datos en formato binario.

open → Se usa para abrir archivos (open("archivo.txt", "r")).

print → Muestra información en la consola.

len → Devuelve la cantidad de elementos en una lista, string, etc.

input → Recibe datos del usuario a través de la consola.

range → Genera una secuencia de números en un rango determinado.

super → Llama a métodos de la clase padre en herencia.

self → Se usa dentro de clases para hacer referencia a la instancia actual del objeto.

type → Devuelve el tipo de una variable (type(5) devuelve <class 'int'>).

enumerate → Función que permite iterar sobre listas o tuplas y obtener tanto el índice como el valor.

zip → Combina dos o más listas en pares de elementos.

map → Aplica una función a todos los elementos de una lista sin necesidad de usar un bucle.

filter → Filtra los elementos de una lista que cumplen una condición dada.

sorted → Ordena los elementos de una lista.

reversed → Invierte una lista o iterador.

sum → Suma todos los elementos de una lista o tupla.

max → Devuelve el valor máximo en una lista.

min → Devuelve el valor mínimo en una lista.

round → Redondea un número decimal a un número específico de decimales.

abs → Devuelve el valor absoluto de un número.

chr → Devuelve el carácter ASCII de un número.

ord → Devuelve el código ASCII de un carácter.

format → Formatea cadenas de texto con variables.

f-string → Otra forma de formatear cadenas de manera más simple.

bin → Convierte un número en su representación binaria.

hex → Convierte un número en su representación hexadecimal.

oct → Convierte un número en su representación octal.

all → Devuelve True si todos los elementos de una lista o tupla son verdaderos.

any → Devuelve True si al menos un elemento de una lista o tupla es verdadero.
