```python
# Ejercicio 1: Clase Estudiante
class Estudiante:
    def __init__(self, nombre, nota, id_estudiante):
        self.__nombre = nombre
        self.__nota = nota if 0 <= nota <= 100 else 0
        self.id_estudiante = id_estudiante

    def get_nombre(self):
        return self.__nombre

    def get_nota(self):
        return self.__nota

    def set_nota(self, nueva_nota):
        if 0 <= nueva_nota <= 100:
            self.__nota = nueva_nota

# Ejercicio 2: Clase CuentaBancaria
class CuentaBancaria:
    def __init__(self, titular, saldo, numero_cuenta):
        self.__titular = titular
        self.__saldo = max(saldo, 0)
        self._numero_cuenta = numero_cuenta

    def depositar(self, monto):
        if monto > 0:
            self.__saldo += monto

    def retirar(self, monto):
        if 0 <= monto <= self.__saldo:
            self.__saldo -= monto

    def get_titular(self):
        return self.__titular

# Ejercicio 3: Clase Libro
class Libro:
    def __init__(self, titulo, autor, paginas, isbn):
        self.__titulo = titulo
        self.__autor = autor
        self._paginas = max(paginas, 1)
        self.isbn = isbn

    def get_titulo(self):
        return self.__titulo

    def get_autor(self):
        return self.__autor

    def set_paginas(self, nuevas_paginas):
        if nuevas_paginas > 0:
            self._paginas = nuevas_paginas

# Ejercicio 4: Clase Vehiculo
class Vehiculo:
    def __init__(self, marca, velocidad, tipo_vehiculo):
        self.__marca = marca
        self.__velocidad = max(0, velocidad)
        self.tipo_vehiculo = tipo_vehiculo

    def acelerar(self, incremento):
        if incremento > 0:
            self.__velocidad = min(self.__velocidad + incremento, 200)

    def frenar(self, decremento):
        if decremento > 0:
            self.__velocidad = max(self.__velocidad - decremento, 0)

# Ejercicio 5: Clase Empleado
class Empleado:
    def __init__(self, nombre, salario, departamento, id_empleado):
        self.__nombre = nombre
        self.__salario = max(0, salario)
        self._departamento = departamento
        self.id_empleado = id_empleado

    def get_nombre(self):
        return self.__nombre

    def get_salario(self):
        return self.__salario

    def get_departamento(self):
        return self._departamento

    def set_salario(self, nuevo_salario):
        if nuevo_salario >= 0:
            self.__salario = nuevo_salario

# Ejercicio 6: Clase Producto
class Producto:
    def __init__(self, nombre, precio, categoria, codigo_producto):
        self.__nombre = nombre
        self.__precio = max(0.01, precio)
        self._categoria = categoria
        self.codigo_producto = codigo_producto

    def get_nombre(self):
        return self.__nombre

    def get_precio(self):
        return self.__precio

    def set_precio(self, nuevo_precio):
        if nuevo_precio > 0:
            self.__precio = nuevo_precio

# Ejercicio 7: Clase Telefono
class Telefono:
    def __init__(self, numero, marca, bateria, modelo):
        self.__numero = numero
        self.__marca = marca
        self._bateria = min(max(bateria, 0), 100)
        self.modelo = modelo

    def cargar_bateria(self, cantidad):
        if cantidad > 0:
            self._bateria = min(self._bateria + cantidad, 100)

    def usar_telefono(self, cantidad):
        if cantidad > 0:
            self._bateria = max(self._bateria - cantidad, 0)

# Ejercicio 8: Clase Paciente
class Paciente:
    def __init__(self, nombre, edad, historial, numero_seguro):
        self.__nombre = nombre
        self.__edad = max(0, edad)
        self._historial = historial
        self.numero_seguro = numero_seguro

    def get_nombre(self):
        return self.__nombre

    def get_edad(self):
        return self.__edad

    def set_edad(self, nueva_edad):
        if nueva_edad >= 0:
            self.__edad = nueva_edad

# Ejercicio 9: Clase Computadora
class Computadora:
    def __init__(self, marca, almacenamiento, sistema_operativo, numero_serie):
        self.__marca = marca
        self.__almacenamiento = max(1, almacenamiento)
        self._sistema_operativo = sistema_operativo
        self.numero_serie = numero_serie

    def get_marca(self):
        return self.__marca

    def get_almacenamiento(self):
        return self.__almacenamiento

    def get_sistema_operativo(self):
        return self._sistema_operativo

    def set_almacenamiento(self, nuevo_almacenamiento):
        if nuevo_almacenamiento > 0:
            self.__almacenamiento = nuevo_almacenamiento

# Ejercicio 10: Clase ReservaHotel
class ReservaHotel:
    def __init__(self, nombre_huesped, noches, tipo_habitacion, codigo_reserva):
        self.__nombre_huesped = nombre_huesped
        self.__noches = max(1, noches)
        self._tipo_habitacion = tipo_habitacion
        self.codigo_reserva = codigo_reserva

    def get_nombre_huesped(self):
        return self.__nombre_huesped

    def get_noches(self):
        return self.__noches

    def set_noches(self, nuevas_noches):
        if nuevas_noches > 0:
            self.__noches = nuevas_noches
```
